```python
import sys
from pathlib import Path
import joblib
import os
import random
import json
from production.preprocess_and_predict import preprocess_user_input
from production.generate_recommendation import generate_sentence

ROOT = Path().resolve().parents[0]
sys.path.append(str(ROOT))
```

```python
def generate_realistic_test_case():
    """
    Generate a single synthetic test case using ranges that match
    the model's real training distribution. This helps avoid bias
    and unrealistic outliers that distort predictions.
    """

    # Randomly assign gender
    gender = random.choice(["F", "M", "Other"])

    # Ranges reflect realistic adult fitness, health, and menstrual cycle values.
    age = random.randint(20, 61)
    height_cm = round(random.uniform(165.3, 189.3), 1)
    weight_kg = round(random.uniform(53.7, 88.0), 1)

    duration_minutes = random.randint(21, 115)
    intensity = random.choice(["Low", "Medium", "High"])
    calories_burned = round(random.uniform(2.6, 19.7), 1)

    hours_sleep = round(random.uniform(5.6, 8.1), 1)
    stress_level = random.randint(1, 10)

    daily_steps = random.randint(5406, 11120)
    hydration_level = round(random.uniform(1.5, 3.4), 1)

    # Only females get period cycle data
    days_since_last_period = random.randint(0, 35) if gender == "F" else None

    
    test_case = {
        "age": age,
        "gender": gender,
        "height_cm": height_cm,
        "weight_kg": weight_kg,
        "duration_minutes": duration_minutes,
        "intensity": intensity,
        "calories_burned": calories_burned,
        "hours_sleep": hours_sleep,
        "stress_level": stress_level,
        "daily_steps": daily_steps,
        "hydration_level": hydration_level,
        "days_since_last_period": days_since_last_period
    }
    return test_case

```

```python
# Load final model
MODEL = joblib.load("../models/rf_activity_87pct.pkl")
```

```python
def predict(user_json):
    """
    Runs the full ML prediction pipeline:
      1. Preprocess user input into model-ready features.
      2. Generate the model's probability predictions.
      3. Create a human-readable sentence based on the prediction.
      4. Return both the raw probabilities and the generated text.
    """

    # Convert raw user JSON into the numerical feature vector expected by the model
    X = preprocess_user_input(user_json)
    print("Preprocessed input:", X)
    print("================================================")

    # Run the model and extract the 1st row (since predict outputs a batch)
    probs = MODEL.predict(X)[0]
    print("Exercise prediction:", probs)

    # Convert model output + user data into a readable explanation
    sentence = generate_sentence(user_json, probs)
    print(sentence)

    return probs, sentence
```

```python
# Create list of test cases
test_cases = [generate_realistic_test_case() for i in range(16)]
```

```python
# Get predictions for each test case
for test in test_cases:
    predict(test)
```
